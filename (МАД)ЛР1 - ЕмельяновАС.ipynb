```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_excel('Датасет.xlsx', header=None)
print(data)
```

```python
plt.boxplot(data)
plt.scatter(np.ones(len(data)), data, s=15, alpha=0.5)
plt.show()
```

```python
plt.hist(data[0], len(data))
plt.title('Частота')
plt.xlabel('мм')
plt.ylabel('Частота')
plt.show()
```

```python
sns_plot = sns.distplot(data[0], 19)
fig = sns_plot.get_figure()
```

```python
# Считаем среднее значение
print(f'mean: {np.mean(data)}')
# Находим моду
print(f'mode: {stats.mode(data, keepdims=True)[0][0]}')
# Находим медиану
print(f'median: {np.median(data)}')
# Находим выборочное среднеквадратичное отклонение
print(f'STD: {np.std(data, ddof=1)}')
```

```python
# Определяем доверительный интервал для генерального среднего значения
print(f'interval: {stats.t.interval(1 - 0.05, len(data) - 1, loc=np.mean(data), scale=stats.sem(data))}')
```

```python
# Находим количество образцов для доверительного интервала ±0,5 мм
print(f'Number of samples: {round(((2.026 * np.std(data, ddof=1)) / 0.5) ** 2)}')
```

```python
# Находим коэффициент вариации
print(f'Relative STD:{100 * np.std(data, ddof=1) / np.mean(data)}')
```
